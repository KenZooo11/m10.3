```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler


mean_100 = 3
std_dev_100 = 1
data_100 = np.random.normal(mean_100, std_dev_100, 100)

mean_10000 = 3
std_dev_10000 = 1
data_10000 = np.random.normal(mean_10000, std_dev_10000, 10000)

print("Dane dla 100 próbek:")
print(data_100)
print("\nDane dla 10 000 próbek:")
print(data_10000)
```

```python
mean_100 = np.mean(data_100)
mean_10000 = np.mean(data_10000)

print("Średnia dla 100 próbek:", mean_100)
print("Średnia dla 10000 próbek:", mean_10000)
print("Dla obu zestawów danych wartości te są podobne.")
```

```python
median_100 = np.median(data_100)
median_10000 = np.median(data_10000)

print("Mediana dla 100 próbek:", median_100)
print("Mediana dla 10000 próbek:", median_10000)
print("Dla obu zestawów danych wartości te również są podobne.")
```

```python
mode_100 = stats.mode(data_100)

print("Moda dla 100 próbek:", mode_100)
print("W przykładzie każda wartość występuje tylko raz dlatego, wyświetlana jest pierwsza, najmniejsza wartość")
plt.figure(figsize=(10, 6))
plt.hist(data_100, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_100, color='red', label = 'Średnia')
plt.axvline(median_100, color='green', label = 'Mediana')
plt.axvline(mode_100[0], color='orange', label = 'Moda')
plt.legend(loc='upper right')
plt.title('Histogram dla 100 próbek')
plt.xlabel('Wartość')
plt.ylabel('Częstość')
plt.grid(True)
plt.show()
```

```python
mode_10000 = stats.mode(data_10000)

print("Moda dla 10000 próbek:", mode_10000)
print("W przykładzie każda wartość występuje tylko raz dlatego, wyświetlana jest pierwsza, najmniejsza wartość. Ze względu na większą liczbę danych dlatego, wartość najmniejsza jest mniejsza niż przy 100 elementach zbioru.")

plt.figure(figsize=(10, 6))
plt.hist(data_10000, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_10000, color='red', label = 'Średnia', alpha=0.5)
plt.axvline(median_10000, color='green', label = 'Mediana')
plt.axvline(mode_10000[0], color='orange', label = 'Moda')
plt.legend(loc='upper right')
plt.title('Histogram dla 10 000 próbek')
plt.xlabel('Wartość')
plt.ylabel('Częstość')
plt.grid(True)
plt.show()
```

```python
percentiles_100 = np.percentile(data_100, [0, 25, 50, 75, 100])

print("Percyntyle dla 100 próbek:", percentiles_100)

plt.figure(figsize=(10, 6))
plt.hist(data_100, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(percentiles_100[0], color='red', linestyle='--', label='0%')
plt.axvline(percentiles_100[1], color='green', linestyle='--', label='25%')
plt.axvline(percentiles_100[2], color='orange', linestyle='--', label='50%')
plt.axvline(percentiles_100[3], color='purple', linestyle='--', label='75%')
plt.axvline(percentiles_100[4], color='black', linestyle='--', label='100%')
plt.xlabel('Wartość')
plt.ylabel('Częstość')
plt.legend()
plt.grid(True)
plt.show()
```

```python
percentiles_10000 = np.percentile(data_10000, [0, 25, 50, 75, 100])

print("Percyntyle dla 10 000 próbek:", percentiles_10000)

print("Na przykładzie widać doskonale jak bardzo najniższa oraz najwyższa wartość w zbiorach liczb się różni. 25.percentyl oraz 75 są zdecydowanie bliżej mediany w zbiorze 10 000 liczb niż 100 liczb.")

plt.figure(figsize=(10, 6))
plt.hist(data_10000, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(percentiles_10000[0], color='red', linestyle='--', label='0%')
plt.axvline(percentiles_10000[1], color='green', linestyle='--', label='25%')
plt.axvline(percentiles_10000[2], color='orange', linestyle='--', label='50%')
plt.axvline(percentiles_10000[3], color='purple', linestyle='--', label='75%')
plt.axvline(percentiles_10000[4], color='black', linestyle='--', label='100%')
plt.xlabel('Wartość')
plt.ylabel('Częstość')
plt.legend()
plt.grid(True)
plt.show()
```

```python
range_100 = np.max(data_100) - np.min(data_100)
range_10000 = np.max(data_10000) - np.min(data_10000)

print("Zakres dla 100 próbek:", range_100)
print("Zakres dla 10000 próbek:", range_10000)
print("W zbiorze danych 10 000 próbek zakres danych jest prawie dwa razy większy, niż zakres w zbiorze 100 próbek.")
```

```python
std_dev_100 = np.std(data_100, ddof=1)
std_dev_10000 = np.std(data_10000, ddof=1)

print("Odchylenie standardowe dla 100 próbek:", std_dev_100)
print("Odchylenie standardowe dla 10000 próbek:", std_dev_10000)
print("Odchylenie standardowe jest niższe dla 10000 próbek, ponieważ dane są mniej rozproszone.")
```

```python
iqr_100 = percentiles_100[3] - percentiles_100[1]

print("Rozstęp międzykwartylowy (IQR) dla 100 próbek:", iqr_100)

plt.figure(figsize=(5, 8))
plt.boxplot(data_100, vert=True, patch_artist=True, medianprops=dict(color='red'))
plt.ylabel('Wartość')
plt.xticks([1], ['Dane'])
plt.grid(True)
plt.scatter([1]*5, percentiles_100, color=['red', 'green', 'orange', 'purple', 'black'], label=['0%', '25%', '50%', '75%', '100%'])
plt.show()
```

```python
iqr_10000 = percentiles_10000[3] - percentiles_10000[1]

print("Rozstęp międzykwartylowy (IQR) dla 10000 próbek:", iqr_10000)
print("Wartość minimalna i maksymalna jest bliżej, dlatego  tematyczne pudełko jest bardziej 'zbite' w liczby. Dużo jest wartości odstających, zatem występuje zupełna odwrotność niż dla 100 próbek, gdzie tych wartości jest zdecydowanie mniej.")
plt.figure(figsize=(5, 8))
plt.boxplot(data_10000, vert=True, patch_artist=True, medianprops=dict(color='red'))
plt.ylabel('Wartość')
plt.xticks([1], ['Dane'])
plt.grid(True)
plt.scatter([1]*5, percentiles_10000, color=['red', 'green', 'orange', 'purple', 'black'], label=['0%', '25%', '50%', '75%', '100%'])
plt.show()
```

```python
variance_100 = np.var(data_100)
variance_10000 = np.var(data_10000)

print("Wariancja dla 100 próbek:", variance_100)
print("Wariancja dla 10000 próbek:", variance_10000)
print("Wariacja dla 10000 próbek jest mniejsza, dlatego że występuje większe zagęszczenie liczb w zbiorze, co za tym idzie odległości między kolejnymi liczbami są mniejsze.")
```

```python
skewness_100 = stats.skew(data_100)
skewness_10000 = stats.skew(data_10000)

print("Asymetria rozkładu dla 100 próbek:", skewness_100)
print("Asymetria rozkładu dla 10000 próbek:", skewness_10000)
print("Asymetria rozkładu dla 100 próbek jest negatywnie dodatnia, dla 10000 próbek wynik jest bliski zeru co wskazuje na symetryczny układ.")
```

```python
scaler_MinMax = MinMaxScaler()
normalized_data_values = scaler_MinMax.fit_transform(data_100.reshape(-1, 1))
plt.hist(normalized_data_values, bins=20)
plt.axvline(x = normalized_data_values.mean(), color='red', label = 'Średnia')
plt.show()
print(f'Średnia: {normalized_data_values.mean()}')
print(f'Odchylenie standardowe: {np.std(normalized_data_values, ddof=1)}')
```
